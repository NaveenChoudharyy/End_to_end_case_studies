###### EDA on titles

```python
titles.head()
```

```python
# Dropping duplicate records

titles = titles.drop_duplicates()
```

```python
titles.info()
```

```python
# Dropping "duration_seasons" as only 1971 not nulls out of 6236.

titles = titles.drop(columns='duration_seasons', axis=1)
```

```python
# Dropping all those records where any  of the record contains null as we cannot process that record


titles = titles.dropna(axis=0)
```

```python
titles.info()
```

```python
titles.nunique()
```

```python
# There are some "title" having more then one "description". Extracting those records for analysistitles.


titles[titles.title.isin(list(titles.loc[titles.title.duplicated(), 'title']))].sort_values('title')

# By analyzing the below table we can conclude that there are many moview which are having same names 
# but either they are relesed in different years or thay as different movies as disciption of the movie 
# is different from one another. In conclustion we can say that all these movies having same names
# are not duplicate records.
```

```python
# Correcting format of "date_added"

titles.date_added = pd.to_datetime(titles.date_added)
```

```python
# Creating two more columns i.e. "year" and "month" 

titles["added_month"] = titles.date_added.dt.month
titles["added_month_name"] = titles.date_added.dt.month_name()
titles["added_year"] = titles.date_added.dt.year


# As we have 'year' and 'month' we can drop variable "date_added".

titles = titles.drop(columns=['date_added'])
```

```python
# "release_year" and "show_id" should be int not float. Correcting this.


titles.release_year = titles.release_year.astype(int)
titles.show_id = titles.show_id.astype(int)

titles
```

```python
# "type" has cardinality as 1, so we can drop this also.

titles = titles.drop(columns='type', axis=1)
```

```python
### Correcting the index sequence in all the files

titles = titles.reset_index().drop(columns='index')
directors = directors.reset_index().drop(columns='index')
countries = countries.reset_index().drop(columns='index')
cast = cast.reset_index().drop(columns='index')
category = category.reset_index().drop(columns='index')
```

```python
# Moreging all the datasets and bringing on the level of show_id

final_data = pd.merge(left=titles, right=directors, 
                      on='show_id').merge(right=countries, 
                                          on='show_id').merge(right=countries,
                                                             on='show_id').merge(right=cast,
                                                                                on='show_id').merge(right=category,
                                                                                                   on='show_id')


final_data.head()
```
