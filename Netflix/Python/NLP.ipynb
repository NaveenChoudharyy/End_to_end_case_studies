```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vector = CountVectorizer(stop_words= list(stop_words)).fit(data.description)

vector
```

```python
vact_col = vector.get_feature_names_out()
```

```python
vector = vector.transform(data.description)
```

```python
vector_matrix = vector.todense()
```

```python
vect_data = pd.DataFrame(data=vector_matrix, columns=vact_col)
```

```python
vect_data = pd.DataFrame(vect_data.sum()).reset_index()
```

```python
vect_data.columns = ['col_name', 'freq']

vect_data.sort_values(by='freq', ascending=False)
```
